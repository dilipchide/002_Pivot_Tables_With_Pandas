# what was the average Height - Numerical data

# How do we do that ?
#We can do that with a pivot table.

```python
df.pivot_table(index='Year',columns='Sport',values='Height').dropna(thresh=30)
```

# what about max and min

```python
df.pivot_table(index='Year',columns='Sport',values='Height',aggfunc='min').dropna(thresh=30)
```

```python
df.pivot_table(index='Year',columns='Sport',values='Height',aggfunc='max').dropna(thresh=30)
```

# What about if we need multiple functions?

```python
df.pivot_table(index='Year',columns='Sport',values='Height',aggfunc=['min','mean','max']).dropna(thresh=30)
```

# Pivot table are veryy powerful, Its good to have two categorical variables and with one numerical vale, Just like you have seen above
